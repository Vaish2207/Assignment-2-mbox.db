```python
cursor.execute('''create table Counts(org TEXT, count INTEGER)''')
```

```python
# Establish database connection
conn = sqlite3.connect('mbox.db')
cursor = conn.cursor()

# Read data from txt file and insert into the table
with open('mbox.txt', 'r', encoding='utf-8') as file:
    for line in file:
        # Split the line by whitespace
        parts = line.strip().split()
        if len(parts) >= 2:
            # Assuming the first part is the organization and the second part is the count
            org = parts[0]
            count = parts[1]
            cursor.execute('''INSERT INTO counts (org, count) VALUES (?, ?)''', (org, count))
        else:
            print("Skipping line with unexpected format:", line.strip())

# Commit changes and close connection
conn.commit()
```

```python
# Re-open the cursor after committing changes
cursor = conn.cursor()

# Execute query to retrieve data
cursor.execute('''SELECT * FROM Counts ORDER BY count DESC''')
rows = cursor.fetchall()

# Display the query result
for row in rows:
    print(row)

# Close connection
conn.close()
```
